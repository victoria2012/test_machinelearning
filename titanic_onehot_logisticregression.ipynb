### 후보군 :  pclass, sex, sibsp, parch, embarked, class, who, adult_male, embark_town, alive, alone

```python
import pandas as pd               
dfx = pd.get_dummies(df['sex']) 
# female -> 0, male ->1                
```

```python
dfx
```

```python
df = pd.concat([df, dfx], axis=1)
df
```

```python
dfx = pd.get_dummies(df['pclass'], prefix='pclass')
df = pd.concat([df,dfx],axis=1)
df
```

```python
# female	male pclass_1	pclass_2	pclass_3
```

# 연속형이 있을 때는 표준정규분포로 바꿔야 한다. (변수에 담아서 진행)

```python
X = df[['fare','female','male','pclass_1','pclass_2','pclass_3']]
Y = df[['survived']]
X.shape, Y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
```

```python
X = scaler.transform(X)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(X, Y)
```

```python
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
```

```python
logR.fit(x_train, y_train)
```

```python
logR.score(x_train, y_train)
```
